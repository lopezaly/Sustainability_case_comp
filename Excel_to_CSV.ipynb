```python
import pandas as pd
```

```python
#read in data from UPC Reference Data.xlsx
energy_star = pd.read_excel("UPC Reference Data.xlsx", sheet_name='EnergyStar', skiprows=5)
sqft = pd.read_excel("UPC Reference Data.xlsx", sheet_name='Room Types Sq Ft', skiprows=3)
ladwp_carbon = pd.read_excel("UPC Reference Data.xlsx", sheet_name='LADWP Carbon Intensity', skiprows=5)
```

```python
#read in data from UPC Utility Data 2018-2021.xlsx
elec = pd.read_excel('UPC Utility Data 2018-2021.xlsx', sheet_name='Electricity', skiprows=5)
nat_gas = pd.read_excel('UPC Utility Data 2018-2021.xlsx', sheet_name='Natural Gas', skiprows=2)
water = pd.read_excel('UPC Utility Data 2018-2021.xlsx', sheet_name='Water', skiprows=4)
chill_water = pd.read_excel('UPC Utility Data 2018-2021.xlsx', sheet_name='Chilled Water', skiprows=4)
```

```python
elec = elec.drop(0)
water = water.drop(0)
```
